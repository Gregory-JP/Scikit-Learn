```python
import pandas as pd
```

```python
df = pd.read_csv('../../BasesDados/plano_saude2.csv')
```

```python
x_planoSaude = df.iloc[:, 0:1].values
x_planoSaude
```

```python
y_planoSaude = df.iloc[:, 1].values
y_planoSaude
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor_RF = RandomForestRegressor(n_estimators=10)
```

```python
regressor_RF.fit(x_planoSaude, y_planoSaude)
```

```python
regressor_RF.score(x_planoSaude, y_planoSaude)
```

```python
import numpy as np
```

```python
x_teste = np.arange(min(x_planoSaude), max(x_planoSaude), step=0.1)
x_teste
```

```python
x_teste = x_teste.reshape(-1, 1)
x_teste.shape
```

```python
import plotly.express as px
```

```python
grafico = px.scatter(x=x_planoSaude.ravel(), y=y_planoSaude)
grafico.add_scatter(x=x_teste.ravel(), y=regressor_RF.predict(x_teste), name='Regressao')
grafico.show()
```
