```python
import pickle
```

```python
with open('../BasesDados/credit.pkl', 'rb') as f:
    x_treino, y_treino, x_teste, y_teste = pickle.load(f)
```

```python
x_treino.shape, y_treino.shape
```

```python
x_teste.shape, y_teste.shape
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
rede_neural = MLPClassifier(max_iter=1500, verbose=True, tol=0.0000100, solver='adam', activation='relu', hidden_layer_sizes=(2, 2))
```

```python
rede_neural.fit(x_treino, y_treino)
```

```python
previsoes = rede_neural.predict(x_teste)
```

```python
previsoes # 0 paga, 1 não paga
```

```python
y_teste # previsao
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_teste, previsoes)
```

```python
print(classification_report(y_teste, previsoes))
```

```python
from yellowbrick.classifier import ConfusionMatrix
```

```python
cm = ConfusionMatrix(rede_neural)
cm.fit(x_treino, y_treino)
cm.score(x_teste, y_teste)
```
